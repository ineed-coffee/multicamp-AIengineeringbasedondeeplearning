```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
filepath = 'D:/downloads/house-prices-advanced-regression-techniques/'
train=pd.read_csv(filepath+"train.csv")
test=pd.read_csv(filepath+"test.csv")
```

```python
train.head()
#test.head()
```

```python
train.isnull().sum()[train.isnull().sum()>0]

train=train.drop(['Alley', 'FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1)
test=test.drop(['Alley', 'FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1)
```

```python
corr=train.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.xticks(rotation=45 , size=7)
plt.yticks(rotation=0 , size=7)
```

```python
corr.SalePrice.sort_values(ascending=False)[1:11]
```

```python
rev=corr.SalePrice[corr.SalePrice>0.5]
rev
rev_cols=list(rev.index.values)
rev_cols

corr2=train[rev_cols].corr()

plt.figure(figsize=(8,8))
sns.heatmap(corr2, annot=True, annot_kws={'size':7})
```

```python
import numpy as np
corr=train.corr()
corrsales=corr[np.abs(corr['SalePrice'])>0.5].index
train=pd.DataFrame(train, columns=corrsales)
corr=train.corr()
corr
```

```python
x = train.drop(['SalePrice'],axis=1).values
y = train['SalePrice'].values
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state = 0)
```

```python
print(len(xtrain))
print(len(ytrain))
print(len(xtest))
print(len(ytest))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_r = RandomForestRegressor(n_estimators=300,random_state=0)
model = rf_r.fit(xtrain,ytrain)
model.score(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
plt.plot(ytest,color='red')
plt.plot(ypred,color='blue')
```

```python
model.score(xtest,ytest)
```

```python
filepath = 'D:/downloads/'
df = pd.read_csv(filepath+'cars.csv')
df.head(1)
```

```python
# 산점도 , 선형적인 패턴 시각적 확인
sns.regplot('speed','dist',data=df)
```

```python
# 박스 플랏 , 극단값 확인
fig ,(ax1,ax2) = plt.subplots(1,2)
sns.boxplot('speed',data=df,ax=ax1,orient='v')
ax1.set_title('car speed')

sns.boxplot('dist',data=df,ax=ax2,orient='v')
ax2.set_title('car dist')
```

```python
# 선형 회귀 분석 : 독립,종속 변수가 정규분포를 따를 때 잘 작동한다.
# 밀도 플랏 : 데이터가 정규분포를 따르는지 시각적 확인

# 박스 플랏 , 극단값 확인
fig ,(ax1,ax2) = plt.subplots(1,2)
sns.kdeplot(df['speed'],ax=ax1)
ax1.set_title('car speed')

sns.kdeplot(df['dist'],ax=ax2)
ax2.set_title('car dist')
```

```python
import scipy.stats

print(scipy.stats.skew(df['speed']))
print(scipy.stats.skew(df['dist']))
```

```python
# 예습 : 회귀분석 함수
from statsmodels.formula.api import ols

res = ols("dist~speed",data=df).fit()
```

```python
res.summary()

'''
R-squared : 모델 설명력 , dist 의 분산을 speed가 약 65% 설명해준다.
Prob (p-value) : 통계적 유의미함 (작을 수록 , p<0.05)
coef : 선형 식의 각 계수들. dist = -17.5791 + 3.9324*speed


결론:
dist 에 대하여 speed 로 예측하는 회귀분석 결과 , 통계적으로 유의미. 근거 -> F-statistic , Prob(p-value)


'''
```

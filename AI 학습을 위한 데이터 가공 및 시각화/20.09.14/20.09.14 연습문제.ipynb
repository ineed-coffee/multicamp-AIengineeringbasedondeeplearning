# 1. iris 데이터셋

### 1.1 :  '종' 열을 one-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
code_2D = ohe.fit_transform(iris.variety.to_numpy().reshape(-1,1)).toarray().astype(np.int64)
code_1D = [''.join(row) for row in code_2D.astype(np.str)]
iris['variety_OHE'] = code_1D
iris.head(3)
```

### 1.2 : 유클리디안 거리 기반 (4.9,3.5,1.4,0.2) -> 종 예측

```python
vect_in = [4.9,3.5,1.4,0.2]
```

```python
def cal_dist(element,comp=vect_in):
    vect_arr = np.array(vect_in)
    diff = element[:4]-vect_arr
    diff = diff.apply(lambda x:x**2)
    return diff.sum()**0.5
```

```python
# 가장 유사한(가까운) 11개 데이터 확인
iris['dist']=iris.apply(cal_dist,axis=1)
Top_11 = iris.sort_values(by='dist').iloc[:11]
Top_11
```

```python
print(f'입력 - {vect_in} 에 대해 종 예측값 : {Top_11.variety.max()}')
```

### __2.1 :타이타닉호 승객을 사망자와 생존자 그룹으로 나누고 각 그룹에 대해 위 라벨별 승객 비율을 구한다.__  
### __(각 그룹별로 비율 전체 합은 1)__  

```python
filepath = 'D:/downloads/'
filename = 'train.csv'
```

```python
titanic = pd.read_csv(filepath+filename)
titanic.head(2)
```

```python
Bins = [1,20,39,50,70,99]
Labels = ['미성년자','청년','중년','장년','노년']
```

```python
# age NaN 값 처리
titanic.Age[titanic.Age.isnull()]=np.where(titanic.Sex=='male',
                                           titanic.Age[titanic.Sex=='male'].mean(),
                                           titanic.Age[titanic.Sex=='female'].mean())
```

```python
titanic['Age_bin']=np.digitize(titanic.Age,bins=Bins)
titanic['Age_label']=titanic['Age_bin'].apply(lambda x:Labels[x-1])
titanic.head(2)
```

```python
sol1 = titanic.groupby(['Survived'])['Age_label'].value_counts(normalize=True)
sol1 = sol1.unstack(level=1)
sol1
```

### 2.2 :생존자 그룹 - 생존자가 가장 많은/적은 카테고리 , 사망자 그룹 - 사망자가 가장 많은/적은 카테고리

```python
max_prop=sol1.apply(lambda x:x.idxmax(),axis=1)
min_prop=sol1.apply(lambda x:x.idxmin(),axis=1)
sol1['비율이 가장 높은 나이층'] = max_prop
sol1['비율이 가장 낮은 나이층'] = min_prop
sol1
```

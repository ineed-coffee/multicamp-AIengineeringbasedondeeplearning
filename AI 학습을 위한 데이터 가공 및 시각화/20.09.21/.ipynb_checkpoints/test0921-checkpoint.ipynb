```python
data = pd.DataFrame({
    'food':['bacon','pork','bacon','beef','Bacon','ham'],
    'ounces':[4,3,12,6,8,3]
})
data
```

```python
meat_to_ani = {'bacon':'pig',
              'pork':'pig',
              'beef':'cow',
              'ham':'pig'}
```

```python
data['animal'] = data.food.apply(lambda x:meat_to_ani[x.lower()])
data
```

시리즈의 경우 apply 와 lambda 를 결합하여 활용할 수도 있지만 , map(딕셔너리) 로 한번에 처리 또한 가능

```python
data = pd.Series([1,-900,2,-999,-1000,3])
data
```

```python
data.replace([-900,900],[np.nan,0]) # 스트링의 replace 와 달리 리스트로 전달, 리스트로 할당이 가능 혹은 한번에 딕셔너리 구조로 전달
```

```python
data = pd.DataFrame(np.arange(12).reshape(3,4),
                   index = ['Seoul','Busan','Daejon'],
                   columns = ['one','two','three','four'])
```

```python
data.index = data.index.map(lambda x :x.upper())
data
```

행 인덱스 혹은 열 인덱스 수정시 rename 메소드 또한 활용 가능

```python
data.rename(index = str.title, columns = str.upper) # 반영하려면 inplace 옵션 True
```

### pd.cut() 구간별 그룹핑

```python
ages = [20,22,25,27,31,21,23,37,61,45,41,32]
bins = [18,25,35,60,100]
cats = pd.cut(ages,bins)
cats
```

Categorical 객체에서 codes = values , categories = index 정도로 이해하면 쉽다.

```python
type(cats)
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages,bins,right=False,labels=['youth','young','middleAged','senior'])
```

```python
data = pd.DataFrame(np.random.randn(1000,4))
data.head()
```

```python
# np.sign 활용
# 절대값이 3 이상이면 3, -3 으로 변경
data[np.abs(data)>3] = np.sign(data)*3
```

```python
data.describe()
```

```python
train['year'] = train.datetime.dt.year
train['month'] = train.datetime.dt.month
train['day'] = train.datetime.dt.day
train['hour'] = train.datetime.dt.hour
train.head(1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib
from matplotlib import font_manager, rc
import platform
if platform.system()=="Windows":
    font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False

import warnings
warnings.filterwarnings("ignore")
```

```python
figure , ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2,ncols=3)
figure.set_size_inches(18,8)
ax1.set(ylabel='Count',title='연도별 대여량')
ax2.set(ylabel='Count',xlabel='month',title='월별 대여량')
ax3.set(ylabel='Count',xlabel='day',title='일별 대여량')
ax4.set(ylabel='Count',xlabel='hour',title='시간별 대여량')

sns.barplot(data=train,x='year',y='count',ax=ax1)
sns.barplot(data=train,x='month',y='count',ax=ax2)
sns.barplot(data=train,x='day',y='count',ax=ax3)
sns.barplot(data=train,x='hour',y='count',ax=ax4)
```

```python
figure , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)
figure.set_size_inches(12,10)
ax1.set(ylabel='Count',title='대여량')
ax2.set(ylabel='Count',xlabel='Season',title='계절별 대여량')
ax3.set(ylabel='Count',xlabel='hour of the day',title='시간별 대여량')
ax4.set(ylabel='Count',xlabel='working day',title='근무일별 대여량')

sns.boxplot(data=train,y='count',orient='v',ax=ax1)
sns.boxplot(data=train,x='season',y='count',ax=ax2)
sns.boxplot(data=train,x='hour',y='count',ax=ax3)
sns.boxplot(data=train,x='workingday',y='count',ax=ax4)
```

```python
train['dayofweek'] = train.datetime.dt.dayofweek
train.dayofweek.value_counts()
```

```python
fig , (ax1,ax2,ax3,ax4,ax5) = plt.subplots(nrows=5)
fig.set_size_inches(18,25)
sns.pointplot(data=train,x='hour',y='count',ax=ax1)
sns.pointplot(data=train,x='hour',y='count',hue='workingday',ax=ax2)
sns.pointplot(data=train,x='hour',y='count',hue='dayofweek',ax=ax3)
sns.pointplot(data=train,x='hour',y='count',hue='weather',ax=ax4)
sns.pointplot(data=train,x='hour',y='count',hue='season',ax=ax5)
```

```python
corrMatt = train[['temp','atemp','casual','registered','humidity','windspeed','count']]
corr_heat = corrMatt.corr()
corr_heat
```

```python
mask = np.array(corr_heat)
mask[np.tril_indices_from(mask)]=False
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corr_heat, square=True, annot=True , vmax = 0.8 , mask = mask)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3)
fig.set_size_inches(12,5)
sns.regplot(x='windspeed',y='count',data=train,ax=ax1)
sns.regplot(x='humidity',y='count',data=train,ax=ax2)
sns.regplot(x='temp',y='count',data=train,ax=ax3)
```

```python
def cym(datetime):
    return f'{datetime.year}-{datetime.month}'


train['ym'] = train['datetime'].apply(cym)
train[['datetime','ym']].head(2)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2)
fig.set_size_inches(18,4)
sns.barplot(data=train,x='year',y='count',ax=ax1)
sns.barplot(data=train,x='month',y='count',ax=ax2)
sns.barplot(data=train,x='year',y='count')

fig2,ax3 = plt.subplots()
fig2.set_size_inches(18,4)
sns.barplot(data=train,x='ym',y='count',ax=ax3)
```

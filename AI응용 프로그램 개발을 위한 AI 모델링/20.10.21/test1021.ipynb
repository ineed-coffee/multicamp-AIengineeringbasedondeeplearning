```python
train = pd.read_csv(path+'basketball_train.csv')
test = pd.read_csv(path+'basketball_test.csv')
df = pd.read_csv(path+'basketball_stat.csv')
```

```python
df.drop(['2P','AST','STL'],axis=1,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(df,test_size=0.2)
```

```python
train.shape,test.shape
```

```python
maxKRange = train.shape[0]//2
Klist = []
for i in range(3,maxKRange,2):
    Klist.append(i)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
xTrain = train[['3P','TRB','BLK']]
yTrain = train['Pos']
```

```python
xTrain.shape,yTrain.shape
```

```python
cv_score =[]
for k in Klist:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn, xTrain, yTrain, cv=10, scoring='accuracy')
    cv_score.append(scores.mean())
cv_score
```

```python
plt.plot(Klist,cv_score)
```

```python
k=Klist[cv_score.index(max(cv_score))]
```

```python
knn = KNeighborsClassifier(n_neighbors=k)

xTrain = train[['3P','BLK']]
yTrain = train['Pos']

knn.fit(xTrain,yTrain)
```

```python
pred = knn.predict(test[['3P','BLK']])
```

```python
test['pred']=pred
test[['Pos','pred']]
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test['Pos'],pred)
```

```python
#3개 특징 모두 활용하여 예측 정확도 출력

knn = KNeighborsClassifier(n_neighbors=k)

xTrain = train.drop(['Pos','Player'],axis=1)
yTrain = train['Pos']

knn.fit(xTrain,yTrain)

pred = knn.predict(test.drop(['Pos','Player'],axis=1))

accuracy_score(test['Pos'],pred)
```

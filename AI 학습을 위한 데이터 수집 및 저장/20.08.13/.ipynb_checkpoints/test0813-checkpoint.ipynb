* __전체 연도별 정보로__  

```python
all_df=[]
for year in range(1880,2011):
    df = pd.read_csv(f'c:/babynames/yob{year}.txt',names=['Names','Sex','Births'])
    df['year']=year
    all_df.append(df)
print(len(all_df))
```

* __여러 DataFrame을 시리얼로 연결 : .concat__

```python
df_concat = pd.concat(all_df)
# 인덱스를 새로 조정하는법 : 
# 1. df_concat = pd.concat(all_df,ignore_index=True)
# 2. df_concat = pd.concat(all_df).reset_index()
```

```python
df_concat.info()
```

* __특정 칼럼으로 피봇팅하여 요약__  

```python
#원하는 포멧
#          F     M
# 1880     ?     ?
# 1881     ?     ?
# ...

# 2010     ?     ?
```

```python
df_concat.pivot_table('Births',aggfunc=sum,index='year',columns='Sex')
```

```python
total_births = df_concat.pivot_table('Births',aggfunc=sum,index='year',columns='Sex')
```

```python
total_births.plot(title='Total births')
```

```python
# year , sex 그룹화

def add_prop(group):
    group['prop'] = (group.Births*100)/group.Births.sum()
    return group

new = df_concat.groupby(['year','Sex']).apply(add_prop)
new
```

```python
new.groupby(['year','Sex']).prop.sum()
```

* __각 연도별/성별에 따른 선호 이름 1000개씩 추출__

```python

def get_top_1000(group):
    return group.sort_values(by='Births',ascending=False)[:1000]

top1000 = new.groupby(['year','Sex']).apply(get_top_1000)
top1000
```

```python
top1000.reset_index(inplace=True,drop=True)
top1000
```

* __이름 유행 분석__

```python
Boys = top1000[top1000.Sex=='M']
Girls = top1000[top1000.Sex=='F']
```

```python
top1000.pivot_table('Births',index='year',columns='Names',aggfunc=sum)
```

==> 131년동안 각 연도별 인기있는 이름 상위 1000개씩을 추출했을때 중복을 제외하면 6868종류의 이름이 있음을 확인

```python
total_births = top1000.pivot_table('Births',index='year',columns='Names',aggfunc=sum)
total_births.info()
```

* __이름 사용 경향에 대한 다양성 측면에서 분석: 흔한 이름은 기피?__  

```python
# 방법1 : 기준 누적비율을 정하여 이름 종류 가지 수를 조사.
# 방법2 : 상위 지정수의 이름들의 비율누적 값을 연도에 따라 조사.
```

```python
#방법1
df = Boys[Boys.year == 2010]
prop_cumsum = df.sort_values(by='Births',ascending=False).prop.cumsum()
prop_cumsum.values.searchsorted(50)
```

```python
prop_cumsum.iloc[116]
```

```python
# how about 1900?
df_ex = Boys[Boys.year == 1900]
Cnt = df_ex.sort_values(by='Births',ascending=False).prop.cumsum().values.searchsorted(50)+1
Cnt
```

```python
def get_quantile_count(group,q=50):
    return group.sort_values(by='Births',ascending=False).prop.cumsum().values.searchsorted(50)+1
```

```python
diversity = top1000.groupby(['year','Sex']).apply(get_quantile_count)
diversity = diversity.unstack('Sex')
diversity.plot()
```

```python
# 방법2
table2 = top1000.pivot_table('prop',index='year',columns='Sex',aggfunc=sum)
table2.plot()
```

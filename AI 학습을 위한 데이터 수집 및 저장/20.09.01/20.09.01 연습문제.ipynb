# __1. 전체 그룹에 대하여 모든 상관계수 계산__  

```python
critics = {

    '레드벨벳': {'택시운전사': 2.5,'겨울왕국': 3.5,
             '리빙라스베가스': 3.0,'넘버3': 3.5,
             '사랑과전쟁': 2.5,'세계대전': 3.0},

    'BTS': {'택시운전사': 1.0,'겨울왕국': 4.5,
            '리빙라스베가스': 0.5,'넘버3': 1.5,
            '사랑과전쟁': 4.5,'세계대전': 5.0},

    '블랙핑크': {'택시운전사': 3.0,'겨울왕국': 3.5,
             '리빙라스베가스': 1.5,'넘버3': 5.0,
             '세계대전': 3.0,'사랑과전쟁': 3.5},

    '소녀시대': {'택시운전사': 2.5,'겨울왕국': 3.0,
             '넘버3': 3.5,'세계대전': 4.0},

    '마마무': {'겨울왕국': 3.5,'리빙라스베가스': 3.0,
            '세계대전': 4.5,'넘버3': 4.0,
            '사랑과전쟁': 2.5},

    '오마이걸': {'택시운전사': 3.0,'겨울왕국': 4.0,
             '리빙라스베가스': 2.0,'넘버3': 3.0,
             '세계대전': 3.5,'사랑과전쟁': 2.0},

    '모모랜드': {'택시운전사': 3.0,'겨울왕국': 4.0,
             '세계대전': 3.0,'넘버3': 5.0,
             '사랑과전쟁': 3.5},

    '우주소녀': {'겨울왕국': 4.5, '사랑과전쟁': 1.0,
             '넘버3': 4.0}

}
```

```python
def simPearson(data,name1,name2):
    sumX,sumY=0,0
    sumPowX,sumPowY=0,0
    sumXY=0
    count=0
    
    for key in data[name1]:
        if key in data[name2]:
            count+=1
            sumX+=data[name1][key]
            sumPowX+=(data[name1][key])**2
            sumY+=data[name2][key]
            sumPowY+=(data[name2][key])**2
            sumXY+=data[name1][key]*data[name2][key]
    ret= (sumXY-((sumX*sumY)/count))/((sumPowX-((sumX**2)/count))*(sumPowY-((sumY**2)/count)))**0.5
    return ret
```

```python
Group=[*critics.keys()]
Pearson_vals = [[['',1] for __ in range(len(Group))] for _ in range(len(Group))]
for i in range(len(Group)):
    for j in range(i,len(Group)):
        Pearson_vals[i][j][0]=Group[j]
        if i==j:
            continue
        Pearson_vals[j][i][0]=Group[i]
        val = simPearson(critics,Group[i],Group[j])
        Pearson_vals[i][j][1]=val
        Pearson_vals[j][i][1]=val

for i in range(len(Group)):
    print(f'{Group[i]} 와 다른 그룹 사이 상관계수 : ')
    for item in Pearson_vals[i]:
        print(*item)
    print('='*50)
```

# __2. 각 그룹별로 가장 높은 상관계수를 가진 그룹 출력__  

```python
for i in range(len(Group)):
    print(f'{Group[i]} 와 가장 높은 상관계수 관계인 그룹 : ')
    name=''
    val=0
    for item in Pearson_vals[i]:
        if abs(item[1])>abs(val) and item[0]!=Group[i]:
            val = item[1]
            name=item[0]
    print(name,f'{val:.3f}',end=' , ')
    print('양의 상관관계' if val>=0 else '음의 상관관계')
    print('='*50)
```

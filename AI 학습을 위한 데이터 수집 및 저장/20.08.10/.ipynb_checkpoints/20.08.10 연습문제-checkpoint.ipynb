## __1. 인스타그램에서 '제주도맛집' 키워드에 해당되는 게시글 50개를 추출하여 results에 저장__

```python
from selenium import webdriver
from bs4 import BeautifulSoup as BS
import time
import re
import pandas as pd
driver = webdriver.Chrome('c:/mychrome/chromedriver.exe')
time.sleep(2)
driver.get('https://www.instagram.com')
time.sleep(2)
```

* __(수동 로그인)__

```python
def insta_search(kw):
    url = 'https://www.instagram.com/explore/tags/'+kw
    return url

def get_content(driver):
    html = driver.page_source
    soup = BS(html,'lxml')
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content=' '
    tags = re.findall('#[^\s]+',content)

    date = soup.select('time._1o9PC.Nzb55')
    date = date[0]['datetime'][:10]
    likes_source = soup.select('button.sqdOP.yWX7d._8A5w5 > span')
    if likes_source:
        likes=likes_source[0].text
    else:
        likes=0
    place = soup.select('div.M30cS')[0].text
    ret = [content,date,likes,place,tags]
    return ret

def moveNext(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(2)
```

* __게시글 선택__

```python
word='제주도맛집'
url = insta_search(word)
driver.get(url)
time.sleep(2.5)
first = driver.find_element_by_css_selector('div._9AhH0')
first.click()
time.sleep(2.5)
```

* __50개 게시글 정보 수집__

```python
results=[]
cnt=50
for i in range(cnt):
    data=get_content(driver)
    results.append(data)
    moveNext(driver)
results_df = pd.DataFrame(results)
```

## __2. 음수가 아닌 수들이 주어졌을 때 그 수들을 이어서 만들 수 있는 가장 큰 수를 구하시오.__  
* 입력 형식 : 공백으로 구분된 한줄의 입력으로 가정
* 출력 형식 : 한 줄에 걸쳐 만들 수 있는 가장 큰 수 출력

```python
def my_sort(arr):
    for i in range(len(arr)-1):
        for j in range(0,len(arr)-1-i):
            if not comp(arr[j],arr[j+1]):
                arr[j],arr[j+1]=arr[j+1],arr[j]
def comp(a,b):
    return True if int(a+b)>=int(b+a) else False
        
if __name__ == '__main__':
    arr = input().rstrip().split()
    my_sort(arr)                     # O(N^2)
    Ans=int(''.join(arr))
    print(Ans)
    
```

# __Web crawl review__  

```python
from selenium import webdriver
from bs4 import BeautifulSoup as BS
import time
import re
driver = webdriver.Chrome('c:/mychrome/chromedriver.exe')

def insta_search(kw):
    url = 'https://www.instagram.com/explore/tags/'+kw
    return url

def get_content(driver):
    html = driver.page_source
    soup = BS(html,'lxml')
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content=' '
    tags = re.findall('#[^\s]+',content)

    date = soup.select('time._1o9PC.Nzb55')
    date = date[0]['datetime'][:10]
    likes_source = soup.select('button.sqdOP.yWX7d._8A5w5 > span')
    if likes_source:
        likes=likes_source[0].text
    else:
        likes=0
    place = soup.select('div.M30cS')[0].text
    ret = [content,date,likes,place,tags]
    return ret

driver.get('https://www.instagram.com')
time.sleep(2)

```

```python
import re
word='제주도맛집'
url = insta_search(word)
driver.get(url)
time.sleep(3)
first = driver.find_element_by_css_selector('div._9AhH0')
first.click()
time.sleep(3)

get_content(driver)
```

```python
def moveNext(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(2)

moveNext(driver)
```

```python
# 오후 연습 문제
# 50개 게시글을 스크랩 하여 DF 로 저장
results=[]

data=get_content(driver)
result.append(data)
moveNext(driver)
```

### __일반화 함수를 이용하여 시군구 open api 호출__

```python
import requests
import pandas as pd

my_key='536468687a6c656536385561645750'

def SOA_get_data(url,service): # SOA : Seoul Open Api
    data_list = None
    try:
        result_dict = requests.get(url).json()
        result_data = result_dict[service]
        if result_data['RESULT']['CODE']=='INFO-000':
            data_list = result_data['row']
    except KeyError:
        pass
    return data_list

def SOA_make_url(key,s_idx,e_idx,service):
    return f'http://openapi.seoul.go.kr:8088/{key}/json/{service}/{s_idx}/{e_idx}/'

sgg_data = SOA_get_data(SOA_make_url(my_key,1,25,'SdeTlSccoSigW'),'SdeTlSccoSigW')
if not sgg_data:
    print('잘못된 서비스명으로 접근')
else:
    sgg_df = pd.DataFrame(data=sgg_data,columns=['SIG_CD','SIG_KOR_NM','LAT','LNG'])
```

```python
sgg_df.columns=['시군구 코드','시군구 한글명','위도','경도']
sgg_df
```

* 오픈 api 로부터 받아온 데이터로 만든 DataFrame 을 Excel 파일로 저장해보기

```python
sgg_df.to_excel('seoul_sgg_list.xlsx',index=False)
```

### * 시군군 데이터 병합

```python
seoul_sgg = pd.read_excel('seoul_sgg_list.xlsx')
#	시군구 코드	시군구 한글명	위도	경도

seoul_starbucks = pd.read_excel('seoul_starbucks_list.xlsx')
#매장명	위도	경도	매장타입	주소	전화번호

# 피봇(pivot) 함수 : 피봇팅 (특정 기준에 따른 데이터 요약)
seoul_sgg.columns=['시군구 코드','시군구 명','위도','경도']
```

```python
# 구 별로 스타벅스 매장 개수 세기
seoul_starbucks_sgg_count = seoul_starbucks.pivot_table(index = '시군구 명',values='매장명',aggfunc='count').rename(columns={'매장명':'스타벅스 매장 수'})
seoul_starbucks_sgg_count=seoul_starbucks_sgg_count.drop(['67'])
```

```python
# 시군구 정보와 합치기
tmp = pd.merge(seoul_sgg,seoul_starbucks_sgg_count,how='left',on='시군구 명')
```

```python
seoul_sgg_pop = pd.read_excel('pop_df.xlsx')
seoul_sgg_pop.columns = ['시군구 명','주민등록인구']
tmp2 = pd.merge(tmp,seoul_sgg_pop,how='left',on='시군구 명')
```

```python
seoul_sgg_busy = pd.read_excel('busy_df.xlsx')
seoul_sgg_busy.columns = ['시군구 명','종사자 수','사업체 수']
seoul_sgg_total = pd.merge(tmp2,seoul_sgg_busy,how='left',on='시군구 명')
seoul_sgg_total
```

```python
seoul_sgg_total.to_excel('seoul_sgg_total.xlsx')
```

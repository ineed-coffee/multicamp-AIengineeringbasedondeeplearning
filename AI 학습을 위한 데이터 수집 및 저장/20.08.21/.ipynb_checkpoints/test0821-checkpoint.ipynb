__수익률 계산 :__  
* GOOG 어제 종가 1만원 , 오늘 1.2만원이라면 -> 하루동안 수익률=20%  
* 수익률 = (현재 가격-어제 종가)/어제 종가  
* if 60일 전에 매수 , 금일 종가에 매도 인 경우도 마찬가지로 (A-B)/B

```python
# 공분산 , 상관계수
```

__𝑉𝑎𝑟𝑖𝑎𝑛𝑐𝑒=  1/(𝑛−1) ∑24_(𝑘=1)^𝑛▒〖𝑥_𝑘−𝑥 ̅ 〗__  
__𝑠𝑡𝑎𝑛𝑑𝑎𝑟𝑑 𝑑𝑒𝑣𝑖𝑎𝑡𝑖𝑜𝑛=√𝑉𝑎𝑟𝑖𝑎𝑛𝑐𝑒__

__Covariance : 2차원 축으로 , X 모양으로 sharp 하면 관계가 높은 두 데이터, O 모양으로 flat 하면 관계가 적은 두 데이터__

__공분산에서 면적으로 분산 계산 시 (x-x_mean)●(y-y_mean) 으로 => Sum 의 절대값이 클수록 관계가 큰 데이터 관계(양관계,음관계)__

__상관 계수(Corelation) : 표준화가 된 두 변수의 공분산__  
__-1<=corelation<=1 , 절대값 기준 0.3 언저리 = 관계 낮음. 0.7 언저리 = 관계 높음__

```python
# AAPL & IBM 공분산 , 상관계수 구해보기

AAPL_mean,IBM_mean,*rest = price.pct_change().mean().values
AAPL_std,IBM_std,*rest2 = price.pct_change().std().values

N = len(price.iloc[0])

Covariance_df = (price.pct_change()['AAPL']-AAPL_mean)*(price.pct_change()['IBM']-IBM_mean)
Covariance = Covariance_df.sum()/(N-1)
print(Covariance)
    
Corelation_df = ((price.pct_change()['AAPL']-AAPL_mean)/AAPL_std)*((price.pct_change()['IBM']-IBM_mean)/IBM_std)
Corelation = Corelation_df.sum()/(N-1)
print(Corelation)
```

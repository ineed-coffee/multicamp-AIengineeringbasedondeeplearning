### __전체 상품에 대해 각 정보 수집하기__  

```python
Prod_info = []

for prod in products:
    info={}
    try:
        info['name'] = prod.select('.prod_info > .prod_name > a')[0].text.strip()
    except IndexError:
        info['name']='Error'
    try:
        info['price'] = int(prod.select('p.price_sect > a > strong')[0].text.strip().replace(',',''))
    except IndexError:
        info['price']='Error'
    try:
        info['spec'] = prod.select('.spec_list')[0].text.strip().split('/')
    except IndexError:
        info['spec']='Error'
    Prod_info.append(info)
Prod_info
```

```python
# 상기 코드 함수화 : 페이지 검색 결과에 대해 상품 정보를 추출하는 함수

def get_prod_items(Prod_info):
    Prod_info = []
    for prod in products:
        info={}
        try:
            info['name'] = prod.select('.prod_info > .prod_name > a')[0].text.strip()
        except IndexError:
            continue
        try:
            info['price'] = int(prod.select('p.price_sect > a > strong')[0].text.strip().replace(',',''))
        except IndexError:
            continue
        try:
            info['spec'] = prod.select('.spec_list')[0].text.strip().split('/')
        except IndexError:
            continue
            
        Prod_info.append(info)
    return Prod_info


```

### __여러 페이지에 대해 각 정보 수집하기__  

```python
#http://search.danawa.com/dsearch.php?query=%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0&volumeType=allvs&page=2&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods
import urllib.parse as up
def get_search_page_url(kw,page_num):
    uni_kw = up.quote(string=kw)
    return f'http://search.danawa.com/dsearch.php?query={uni_kw}&volumeType=allvs&page={page_num}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods'

print(get_search_page_url('노트북',3))
```

```python
import time
from tqdm import notebook  # 상태 표시, 프로그래스 바 생성에 쓰는 패키지

Tbook = notebook.tqdm
t=10

for p in Tbook(range(1,t+1)):
    time.sleep(1)
```

```python
keyword = '무선청소기'
total_page = 10

total_data=[]
for page in Tbook(range(1,total_page+1)):
    url=get_search_page_url(keyword,page)
    driver.get(url)
    time.sleep(2)
    
    html = driver.page_source
    soup=bs(html,'html.parser')
    prod_items = soup.select('div.main_prodlist.main_prodlist_list > ul.product_list > li.prod_item')
    prod_data = get_prod_items(prod_items)
    total_data+=prod_data
    
# for idx,data in enumerate(total_data):
#     print(f'{idx+1} 번째 페이지 검색결과')
#     print('='*25)
#     print(data)
#     print()
```

```python
# 잘 만들어졌는지 불러와서 확인

test_read = pd.read_excel('danawa_crawling_result.xlsx')
test_read
```

```python
# 회사명과 제품명 구분

titles = test_read['상품명']
Company,Model = [],[]
for title in titles:
    company,model = title.split(' ',1)
    Company.append(company)
    Model.append(model)

test_read['회사명']=Company
test_read['제품명']=Model
test_read.drop('상품명',axis=1)
test_read = test_read[['회사명','제품명','가격','상품 스펙']]
test_read
```

```python
spec_list = test_read['상품 스펙']
category_list,use_time_list,suction_list=[],[],[]
for spec in spec_list:
    spec_l = spec
    spec_l.replace('[','')
    spec_l.replace(']','')
    spec_l = spec_l.split(',')
    t,c,s='0',None,'0'
    for item in spec_l:
        if '사용' in item:
            t = item.split(' ')[-2]
        elif '청소기' in item:
            c = item.replace('[','') 
        elif '흡입력' in item:
            tmp=item.replace('흡입력','')
            tmp=tmp.replace(':','')
            tmp=tmp.replace("'",'')
            s = tmp.strip()
            
    use_time_list.append(t)
    category_list.append(c)
    suction_list.append(s)

for i in range(len(spec_list)):
    print(use_time_list[i],end=' , ')
    print(category_list[i],end=' , ')
    print(suction_list[i])
```

```python
def convert_time_minute(time_in):
    time_out=0
    if '시간' in time_in:
        h,m = time_in.split('시간')
        time_out+=60*int(h)
        if not m:
            return time_out
    else:
        m=time_in
    mm = m.split('분')
    time_out+=int(mm[0])
    return time_out

times=['40분','4분','1시간','3시간30분']
for time in times:
    converted_time = convert_time_minute(time)
    print(f'time = {converted_time}')
```

```python
converted_use_time_list = [convert_time_minute(x) for x in use_time_list]
converted_use_time_list
```

```python
df = pd.DataFrame()
df['회사명'] = Company
df['제품'] = Model
df['카테고리'] = category_list
df['가격'] = test_read['가격']
df['사용시간'] = converted_use_time_list
df['흡입력'] = suction_list
df.head()
```

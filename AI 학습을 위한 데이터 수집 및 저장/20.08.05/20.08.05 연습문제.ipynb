# __#1. 4개의 직사각형이 차지하는 면적__

```python
Rects = [[*map(int,input().split())] for _ in range(4)]
Ans=0
Cords={}
for rect in Rects:
    x_ll,y_ll,x_hr,y_hr = rect
    for x in range(x_ll,x_hr):
        for y in range(y_ll,y_hr):
            try:
                if Cords[(x,y)]:
                    continue            
            except KeyError:
                Cords[(x,y)]=True
                Ans+=1
print(Ans)
            
    
```

# __#2. '하늘과 바람과 별과 시' 수록 시 출력__

```python
from bs4 import BeautifulSoup as BS
import urllib.request as UR
import re

url = 'https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC'
response = UR.urlopen(url)
soup=BS(response,'html.parser')
selector = '#mw-content-text > div > ul:nth-child(6) > li > ul > li > a'
title_objs = soup.select(selector)
print('하늘과 바람과 별과 시')
print('---------------------')
for idx,title in enumerate(title_objs):
    print(f'{idx+1}. {title.string}')
```

# __#3. Stackoverflow 파이썬 질문 제목 출력__

```python
from bs4 import BeautifulSoup as BS
import urllib.request as UR
import re

page_size =15
url = 'https://stackoverflow.com/questions/tagged/python?tab=newest&page=1&pagesize=15'
response = UR.urlopen(url)
soup=BS(response,'html.parser')
selector='#questions div.question-summary div.summary > h3 > a'
title_objs = soup.select(selector)

print('Questions about Python Programming')
print('----------------------------------')
for idx,title in enumerate(title_objs):
    print(f'{idx+1}. {title.string}')
```

# __#4. '다음'에 있는 영화 댓글 추출__

```python
from bs4 import BeautifulSoup as BS
import urllib.request as UR
import re

url_fix = 'https://movie.daum.net/moviedb/grade?movieId=127897&type=netizen&page='
page=1
while True:
    url=url_fix+str(page)
    response = UR.urlopen(url)
    soap = BS(response,'html.parser')
    com_objs = soap.select('#mArticle > div.detail_movie.detail_rating > div.movie_detail > div.main_detail > ul > li > div > p.desc_review')
    if not com_objs:
        break
    for i in com_objs:
        tmp = i.text.strip()
        tmp.replace('\n','')
        tmp.replace('<br>','')
        print(tmp)
        print('---------------')
    page+=1
        
```

# __웹 크롤링,스크래핑__

```python
# 웹사이트(http,ftp) 데이터 추출에 활용하는 라이브러리 : urllub
# urllib.request 모듈 : 웹에 있는 데이터에 접근
# urllib.request 모듈에서 retrieve 메소드 활용
import urllib.request
image_url ='https://www.all-con.co.kr/img/main_logo1.jpg'
file_name = 'test.jpg'
urllib.request.urlretrieve(image_url,file_name)
print('저장완료')
```

__매개변수 값에 따른 웹페이지 추출__

```python
import urllib.parse as up

url='http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'
myarg={'stnId':'184'} # 딕셔너리 형태로 매개변수 지정

params = up.urlencode(myarg)
full_url = url+'?'+params
print(full_url)
res = urllib.request.urlopen(full_url).read()
res_text = res.decode("utf-8")
print(res_text)
```

# __Beautiful Soup 객체 사용__

```python
from bs4 import BeautifulSoup as BS

html='''
<html><body>
<h1 id='title'>스크래핑?</h1>
<p id='body'>웹 페이지 분석</p>
<p>원하는 부분을 추출</p>
</body></html>
'''
```

__id 태그로 find__

```python
soup = BS(html,'html.parser') # parser : 구문 분석
print(soup.find(id='title').string)
print(soup.find(id='body').string)
```

__a 태그로 find , find_all__

```python
html = '''
<html><body>
<ul>
<li><a href='http://www.naver.com'>naver</a></li>
<li><a href='http://www.daum.net'>daum</a></li>
</ul>
</body></html>
'''
```

```python
soup=BS(html,'html.parser')
links = soup.find_all('a')
for link in links:
    print(f'text = {link.string}')
    print(f"attribute = {link.attrs['href']}")
```

__CSS 스타일시트를 활용한 크롤링__

```python
html='''
<html><body>
<div id='books'>
    <h1>프로그래밍</h1>
    <ul class='name'>
        <li>파이썬</li>
        <li>머신러닝</li>
        <li>데이터분석</li>
    </ul>
</div>
</body></html>
'''
```

```python
soup = BS(html,'html.parser')
soup.find('h1').string        # h1 태그가 1개뿐이기에 바로 검색이 된다.
```

```python
soup.select_one('div#books > h1').string
# 부모 태그를 > 로 명시해주고 #으로 아이디 속성 값을 명시해주면 보다 빠른 검색 가능
```

```python
i = soup.select('div#books > ul.name > li')
print(i)
for j in i:
    print(j.string)
# 여러개의 > 활용 가능하고 . 으로 클래스 속성 값 기입
```

- http://finance.yahoo.com

```python
url = 'http://finance.yahoo.com'
res = req.urlopen(url)
soup = BS(res,'html.parser')

#marketsummary-itm-NQ\=F > h3 > span
res2 = soup.select_one('#marketsummary-itm-NQ\=F > h3 > span')
print(res2.string)
```

- ko.wikisource.org/wiki

```python
url='https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC'
# 문자를 컴퓨터가 처리할 수 있는 인코딩 방식 중 유니코드로 표기 됨 : 아스키코드(7bit) , 유니코드()
selector='#mw-content-text > div > ul > li > b > a'

res = req.urlopen(url)
soup = BS(res,'html.parser')
plist = soup.select(selector)
for p in plist:
    print(p.string)
```

__css의 다양한 선택자__

```python
html='''
<ul id='lang'>
    <li id='ja'>java</li>
    <li id='py'>python</li>
    <li id='cpp'>c++</li>
    <li id='sp'>spark</li>
    <li id='ba'>basic</li>
</ul>
'''
```

```python
soup = BS(html,'html.parser')
plist = soup.select('li')
print(plist)
# '파이썬 글자 뽑아내기'

#1. 
soup.select('li')[1].string
soup.find_all('li')[1].string

#2.
soup.select_one('#py').string

#3.
soup.select_one('li#py').string

#4.
soup.select_one('ul > li#py').string

#5.
soup.select_one('ul#lang li#py').string

#6.
soup.select_one('#lang #py').string

#7.
soup.select_one("li[id='py']").string

#8.
soup.select_one('li:nth-of-type(2)').string
```

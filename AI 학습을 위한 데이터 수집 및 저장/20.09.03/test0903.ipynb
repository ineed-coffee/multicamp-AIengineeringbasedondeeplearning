```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filepath='D:/downloads/3405_6663_bundle_archive/'
filename='tmdb_5000_movies.csv'
data=pd.read_csv(filepath+filename)

data.head(2)
```

```python
# 필요로 하는 컬럼만 추출 (다른 데이터 전처리로 표준/정규화 , 문자데이터 형태소분석, 불용어제거, 대/소문자 통일 ..)

data=data[['id','genres','vote_average','vote_count','popularity','title','keywords','overview']]
data['vote_count']
```

```python
data.info()
```

```python
tmp_m=data['vote_count'].quantile(0.90)
tmp_data=data.copy()[data['vote_count']>=tmp_m]
tmp_data.shape
```

```python
tmp_data.head(2)
```

```python
C = tmp_data['vote_average'].mean()
M = tmp_m
```

(WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C

```python
def weighted_rating(x,m=M,c=C):
    v=x['vote_count']
    R=x['vote_average']
    
    return (v/(v+m)*R) + (m/(m+v)*C)

tmp_data['score'] = tmp_data.apply(weighted_rating,axis=1)
tmp_data.head(2)
```

```python
import ast

listStr = '[1,2,3,4,5]'
listData = ast.literal_eval(listStr)
listData
```

```python
tmp_data['genres']=tmp_data['genres'].apply(ast.literal_eval)
tmp_data['keywords']=tmp_data['keywords'].apply(ast.literal_eval)
```

```python
tmp_data[['genres','keywords']].head(3)
```

```python
def get_names(element):
    return [info['name'] for info in element]

tmp_data['genres']=tmp_data['genres'].apply(get_names)
tmp_data['keywords']=tmp_data['keywords'].apply(get_names)
tmp_data[['genres','keywords']].head(2)
```

__내장 eval 함수 : 보안 취약__  
__ast 모듈 literal_eval 함수 : 범위가 제한적인(일반적인 수칙연산 지원 X) 대신 안정적(디렉토리 구조 노출 X, 기본 자료형 지원)__  

```python
def join_names(element):
    return ' '.join(element)

tmp_data['genres']=tmp_data['genres'].apply(join_names)
tmp_data['keywords']=tmp_data['keywords'].apply(join_names)
tmp_data[['genres','keywords']].head(2)
```

```python
tmp_data.to_csv('pre_tmdb_5000_movies.csv',index=False)
print('Done')
```

## __BOW : Bag of words , 문서를 구성하는 모든 단어들에 대해 표시하는 방법__  
  
문서1: 나는 대한민국 사람  
문서2: 나는 미국 사람  
문서3: 나는 중국 사람  
* 코퍼스(corpus, 말뭉치) : {'나는','대한민국','사람','미국','중국'}  
---
__※코퍼스 : 단어(도메인) 사전 --> BOW 는 코퍼스의 부분집합 이라고 생각하면 됨__  
토지소설 -> 원고 -> 토지 코퍼스:토지 소설에 등장하는 단어들  
태백산맥 소설 -> 원고 -> 태백산맥 단어들 중에는 토지 코퍼스에 있는 단어도 있지만 없는 단어도 존재 (ex. '지리산')  
이럴 경우 _(OOV)_ 라고 함 (Out Of Vocab)  
***
__※BOW : 전체문서{d1,d2,..,dn}를 구성하는 단어장(vocabulary,{t1,t2,..,tm})을 만든 다음,__  
__각각의 문서(n개)에 대해 단어장에 있는 각각의 단어가 문서 내에 포함되어 있는지 여부를 표시하는 방법__  


```python
from sklearn.feature_extraction import DictVectorizer
# DictVectorizer 단어의 빈도정보가 담긴 딕셔너리로 부터 BOW 벡터 생성
v=DictVectorizer(sparse=False)
D=[{'A':1,'B':2},{'B':3,'C':1}] # 단어의 빈도 정보가 담긴 딕셔너리 , 2개의 문서 , 3개의 단어
x=v.fit_transform(D)
x
```

* Dense matrix : 요소에 0 이 거의 없는 행렬  
* Sparse matrix : 요소에 0 이 많이 있는 행렬 

```python
v.feature_names_
```

```python
v.transform({'C':4,'D':2}) # 함수 내부 딕셔너리를 v 변수에 저장된 벡터화 방식으로 변환(bow)
```

CountVectorizer  
문서->단어 토큰 리스트->단어 출현 빈도 -> bow 벡터로 변환\

```python
corpus=[ # 해당 도메인에 대한 관련 문서들의 집합
    'This is the first document',
    'This is the second second document',
    'And the third one',
    'The last document'
]

cv = CountVectorizer()
cv.fit(corpus)
print(cv.vocabulary_) # 0~9 까지 총 10개의 단어로 구성
print(cv.transform(corpus).toarray())
```

```python
corpus=[ # 해당 도메인에 대한 관련 문서들의 집합
    'This is the first document',
    'This is the second second document',
    'And the third one',
    'The last document'
]

cv2=CountVectorizer(stop_words=['is','the'])
# stop-_words : 불용어() , 리스트로 전달
cv2.fit(corpus)
print(cv2.vocabulary_) # 0~9 까지 총 10개의 단어로 구성
print(cv2.transform(corpus).toarray())
```

```python
cv=CountVectorizer(stop_words='english')
cv.fit(corpus)
cv.vocabulary_
```

```python
cv.get_stop_words()
```

```python
cv=CountVectorizer(analyzer='char') # 형태소 분석 : 각 글자마다
cv.fit(corpus)
cv.vocabulary_
```

```python
cv=CountVectorizer(token_pattern='t\w+') # 정규식 활용으로 토큰 사전 생성
cv.fit(corpus)
cv.vocabulary_
```

```python
#n-gram (Word2Vec): 단어의 크기
# 1(모노) 그램 : 토큰 하나만 단어 , 2(바이) 그램 : 두개의 토큰을 하나의 단어로 사용
# This is the first document -> 2-그램으로 표현
# This is, is the, the first, first document
```

```python
cv=CountVectorizer(ngram_range=(1,2)) # 정규식 활용으로 토큰 사전 생성 1~2 그램 범위 전부 출력
cv.fit(corpus)
cv.vocabulary_
```

```python
cv=CountVectorizer(min_df=2) # 빈도수 범위 지정
cv.fit(corpus)
cv.vocabulary_
```

```python
cv.transform(corpus).toarray().sum(axis=0)
# 각 단어별로 빈도수 출력
```

# 1.정규화 각 열(바울, 할로윈) 에 대해 정규화 작업 수행

```python
critics={
    'BTS':{'암수살인':5, '바울':4, '할로윈':1.5},
    '손흥민':{'바울':5, '할로윈':2},
    '레드벨벳':{'암수살인':2.5, '바울':2, '할로윈':1},
    '트와이스':{'암수살인':3.5, '바울':4, '할로윈':5}
}
```

```python
import pandas as pd

c_df = pd.DataFrame(critics)
c_df
```

```python
c_df_reg = pd.DataFrame([(c_df[col]-c_df.min(axis=1))/(c_df.max(axis=1)-c_df.min(axis=1)) for col in c_df],index=c_df.columns)
c_df_reg
```

# 2. 소녀시대와 가장 유사도가 높은 유저

```python
critics2 = {
    '레드벨벳': {'택시운전사': 2.5,'겨울왕국': 3.5,
             '리빙라스베가스': 3.0,'넘버3': 3.5,
             '사랑과전쟁': 2.5,'세계대전': 3.0},
    'BTS': {'택시운전사': 1.0,'겨울왕국': 4.5,
            '리빙라스베가스': 0.5,'넘버3': 1.5,
            '사랑과전쟁': 4.5,'세계대전': 5.0},
    '블랙핑크': {'택시운전사': 3.0,'겨울왕국': 3.5,
             '리빙라스베가스': 1.5,'넘버3': 5.0,
             '세계대전': 3.0,'사랑과전쟁': 3.5},
    '소녀시대': {'택시운전사': 2.5,'겨울왕국': 3.0,
             '넘버3': 3.5,'세계대전': 4.0},
    '마마무': {'겨울왕국': 3.5,'리빙라스베가스': 3.0,
            '세계대전': 4.5,'넘버3': 4.0,
            '사랑과전쟁': 2.5},
    '오마이걸': {'택시운전사': 3.0,'겨울왕국': 4.0,
             '리빙라스베가스': 2.0,'넘버3': 3.0,
             '세계대전': 3.5,'사랑과전쟁': 2.0},
    '모모랜드': {'택시운전사': 3.0,'겨울왕국': 4.0,
             '세계대전': 3.0,'넘버3': 5.0,
             '사랑과전쟁': 3.5},
    '우주소녀': {'겨울왕국': 4.5, '사랑과전쟁': 1.0,
             '넘버3': 4.0}
}
```

```python
def cal_dist(piv_ser,comp_ser):
    ret_ser= piv_ser-comp_ser
    ret_ser= ret_ser*ret_ser
    ret = ret_ser.sum()**0.5
    return ret
```

```python
c_df2 = pd.DataFrame(critics2)
c_df2_reg = pd.DataFrame([(c_df2[col]-c_df2.min(axis=1))/(c_df2.max(axis=1)-c_df2.min(axis=1)) for col in c_df2],index=c_df2.columns)
c_df2_reg
```

```python
piv='소녀시대'
Rel_list
for idx in c_df2_reg.index:
    if idx != piv :
        dist=cal_dist(c_df2_reg.loc[piv],c_df2_reg.loc[idx])
        Rel_list.append((idx,1/(dist+1)))

print('전체 유사도')
Rel_list.sort(key=lambda x:x[-1])
for item in Rel_list:
    print(item)
print('='*50)
print('가장 유사한 그룹')
print(Rel_list[0][0])
```

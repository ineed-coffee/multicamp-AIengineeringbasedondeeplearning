# 1.name 컬럼에서 정규 표현식 활용 호칭 추출 (Mr. , Mrs. , ... , Dr. , Sir, ...)  
### Mr : 0 , Mrs:1 , 나머지 :2 변환 후 gender 칼럼을 새로 생성

```python
pattern = re.compile('[a-zA-Z]+[.]')

def get_call(element):
    ret = pattern.search(element)
    if not ret: return np.nan
    
    tag=ret.group()
    if tag == 'Mr.':
        return 0
    elif tag == 'Mrs.':
        return 1
    else :
        return 2

df['gender'] = df['Name'].apply(get_call)
df.head(10)
```

# 2.age 열에 대해서 빈칸처리 되어 있는 부분 -> 각 해당 데이터의 성별을 토대로  
# 해당 성별(male,female) 평균 값으로 대체

```python
M_mean = round(df[df['Sex']=='male']['Age'].mean())
F_mean = round(df[df['Sex']=='female']['Age'].mean())
print(f'남성 연령 평균 : {M_mean}')
print(f'여성 연령 평균 : {F_mean}')
```

```python
df['Age']=pd.Series(np.where(df['Age'].isnull(),np.where(df['Sex']=='male',M_mean,F_mean),df['Age']))
df.head(10)
```

# 7.age 열에 대해서 value_counts/cut 함수를 이용하여  
# 구간을 나누고 데이터 개수를 세어 열 생성 => 최종: 연령대별 생존/사망률

```python
lastSeg = int(10*(1+df['Age'].max()//10))
segList = list(range(0,lastSeg+1,20))
```

```python
ageSeg_value_cnt = df['Age'].value_counts(bins=segList)
ageSeg_value_cnt
```

```python
ageSeg_cut = pd.cut(df['Age'],bins=segList)
df['Age_segment']=ageSeg_cut
ageSeg_cut
```

```python
age_group = df.groupby('Age_segment')
```

```python
def cal_rate(group_iter):
    M = len(group_iter[1])
    D = len(group_iter[1][group_iter[1]['Survived']==1])
    return D/M

sd_rate_by_age = pd.DataFrame()
sd_rate_by_age['Survival rate'] = pd.Series([cal_rate(group) for group in age_group],index=[group[0] for group in age_group])
sd_rate_by_age['Death rate'] = 1-sd_rate_by_age['Survival rate']
sd_rate_by_age
```

# 10.성별에 따른 생존율/사망률 구하기

```python
Gender_sd_rate = pd.DataFrame(index=df['Sex'].value_counts().index)
Gender_sd_rate['사망률']= [len(group[1][group[1]['Survived']==0])/len(group[1]) for group in df.groupby('Sex')]
Gender_sd_rate['생존율']=1-Gender_sd_rate['사망률']
Gender_sd_rate
```

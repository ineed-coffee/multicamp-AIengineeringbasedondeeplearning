# 6. 어느 회사의 전반기 (1-6월) 실적을 나타내는 데이터프레임과  
# 후반기(7-12월) 실적을 나타내는 데이터프레임을 임의로 만든 뒤 합친다.  
* 실적 정보는 '매출','비용','이익'으로 이루어짐 (이익=매출-비용) , 또한 1년간의 총 실적을 마지막 행으로 추가

```python
first_half = pd.DataFrame({'월':np.arange(1,7),
                          '매출':np.random.randint(10**2,10**6,(6,)),
                          '비용':np.random.randint(10**2,10**6,(6,))})
first_half['이익'] = first_half['매출']-first_half['비용']

second_half = pd.DataFrame({'월':np.arange(7,13),
                          '매출':np.random.randint(10**2,10**6,(6,)),
                          '비용':np.random.randint(10**2,10**6,(6,))})
second_half['이익'] = first_half['매출']-first_half['비용']
```

```python
first_half
```

```python
second_half
```

```python
whole_year = pd.merge(first_half,second_half,on=[*first_half.columns],how='outer')
tot_row = whole_year.sum()
tot_row['월']='총실적'
whole_year.append(tot_row,ignore_index=True)
#whole_year
```

# __오후__

```python
df1=pd.DataFrame(np.arange(6).reshape(3,2),
            index=['a','b','c'],
            columns=['one','two'])

df2=pd.DataFrame(np.arange(4).reshape(2,2)+5,
            index=['a','c'],
            columns=['three','four'])
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],axis=1,keys=['level1','level2'])
```

```python
pd.concat({'level1':df1,'level2':df2},axis=1)  # obj 인자와 keys 인자를 딕셔너리 형태로 한번에 전달가능
```

```python
pd.concat({'level1':df1,'level2':df2},axis=1,names=['L1','L2'])
```

```python
# 데이터 합칠때 (merge,concat)

a=pd.Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],
         index=['f','e','d','c','b','a'])
b=pd.Series(np.arange(len(a),dtype=np.float64),
         index=['f','e','d','c','b','a'])
print(a)
print(b)
```

```python
# a와b를 합친 결과 출력
#1. array[0,3.5,2.0,6.5,8.5,5.0]
#2. array[]

np.array(a.fillna(0)+b)
#np.where(a.isnull(),b,a+b)
```

```python
dfleft = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})

print(dfleft)
print(dfright)
```

```python
pd.merge(dfleft,dfright,how='outer')
```

```python
pd.merge(dfleft,dfright,how='right')
```

```python
dfleft=pd.DataFrame({'KEY':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':[0.5,2.1,3.1,0.3]})
dfleft
```

```python
dfright=pd.DataFrame({'KEY':['K2','K3','K4','K5'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
dfright
```

```python
dfall=pd.merge(dfleft,dfright,how='outer')
dfall
```

```python
dfall.isnull()
```

```python
# 실습 : dfall 의 0,1 행의 A,B 칼럼에 None 할당

dfall.loc[[0,1],['A','B']]=None
dfall
```

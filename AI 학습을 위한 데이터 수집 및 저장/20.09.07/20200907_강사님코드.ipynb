```python
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})
dfright = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
print(dfleft)
print(dfright)
```

```python
pd.merge(dfleft,dfright)
```

```python
pd.merge(dfleft,dfright, how='inner') #키가 동일한 것끼리 merge, default
 #공통으로 있는 KEY 열에 대한 교집합
```

```python
pd.merge(dfleft,dfright, how='outer') #공통으로 있는 KEY 열에 대한 합집합
```

```python
pd.merge(dfleft,dfright) #how='inner', on='KEY' 생략
```

```python
print(dfleft)
print(dfright)
```

```python
pd.merge(dfleft,dfright, how='right') #right outer join
```

```python
print(dfleft)
print(dfright)
pd.merge(dfleft,dfright, how='left', on='KEY') 
```

```python
pd.merge(dfleft,dfright,on='KEY', how='outer', indicator=True) #부가정보
```

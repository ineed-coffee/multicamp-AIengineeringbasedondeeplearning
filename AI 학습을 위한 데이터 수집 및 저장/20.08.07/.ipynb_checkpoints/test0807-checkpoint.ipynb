 - __이미지 크롤링__
    - 검색어 입력 ? 자동차
    - 크롤링할 데이터 개수 :30
    - 자동차 이미지 상위 30개 다운로드 및 특정 경로에 저장

```python
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup as BS
base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
word = input('검색어 : ')
num = int(input('크롤링할 데이터 개수: '))
url = base_url+quote_plus(word)
response = urlopen(url)
soap=BS(response,'html.parser')
images = soap.select('._img')
```

```python
n=1
for image in images:
    imgurl = image['data-source']
    img = urlopen(imgurl).read()
    f = open('image_container/down'+str(n)+'.jpg','wb')
    f.write(img)
    f.close()
    n+=1
    if n>50:
        break
```

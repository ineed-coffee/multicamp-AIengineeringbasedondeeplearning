# 2. 다나와-> 노트북 페이지
### * 가격  
### * 화면
### * CPU
### * 코어
### * SSD
### * RAM
### * 제조사

```python
from bs4 import BeautifulSoup as BS
from urllib.request import urlopen
from selenium import webdriver
import re

driver = webdriver.Chrome('c:/mychrome/chromedriver.exe')
url = 'http://prod.danawa.com/list/?cate=112758&15main_11_02'
driver.get(url)
html = driver.page_source
soap = BS(html,'html.parser')

Laps = soap.select('.prod_main_info')
```

```python
Results=[]
for lap in Laps:
    spec = lap.text
    tmp={}
    price = re.search('[\d,]+0원',spec)
    cpu = re.search('[-\diㄱ-힣]+세대',spec)
    core = re.search('[ㄱ-힣]+코어',spec)
    ssd = re.search('SSD:[\d ]+GB|SSD:[\d ]+TB',spec)
    ram = re.search('DDR4 /[\d ]+GB|\(온보드\) /[\d ]+GB',spec)
#     print(price)
#     print(cpu)
#     print(core)
#     print(ssd)
#     print(ram)
    if price: tmp['price']=price.group()
    if cpu: tmp['cpu']=cpu.group()
    if core: tmp['core']=core.group()
    if ssd: tmp['ssd']=ssd.group();tmp['ssd']=tmp['ssd'].split()[-1]
    if ram: tmp['ram']=ram.group();tmp['ram']=tmp['ram'].split()[-1]
    Results.append(tmp)
```

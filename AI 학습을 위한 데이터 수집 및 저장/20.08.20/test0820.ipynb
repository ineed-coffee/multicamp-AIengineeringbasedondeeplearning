```python
df = pd.DataFrame(np.random.randn(4,3),
            columns=list('bde'),
            index=['Seoul','Busan','Daegu','Incheon'])

df
```

```python
# 각 열에 대해 최대-최소 값 추출

for col in df.columns:
    Mx=df[col].max()
    Mn=df[col].min()
    print(f'{col}열로부터 추출한 최대-최소 : {Mx-Mn}')
```

__apply : 각 컬럼에 해당되는 1차원 배열에 대해 함수를 적용__

```python
f = lambda x:x.max()-x.min()
df.apply(f) # == .apply(f,axis=0) , .apply(f,axis='index')
```

```python
df.apply(f,axis=1)
```

```python
# df 에 저장되어 있는 데이터에 대해 도시 단위 평균 값 

f2 = lambda x: x.mean()
df.apply(f2,axis=1)
```

```python
def f3(x):
    return pd.Series([x.min(),x.max()],index=['min','max'])

df.apply(f3,axis=1)
```

```python
# df 에 있는 모든 수치 데이터에 대해 소수점 이하 2자리까지만 출력

f4 = lambda x: pd.Series(map(lambda x_:round(x_,2),x),index=x.index)
df.apply(f4)

# 열이나 행 단위가 아니라 원소 하나 하나 적용시키는 함수 : applymap

f5 = lambda x: float(f'{x:.2f}')
#df.applymap(f5)
df.applymap(lambda x:round(x,2))
```

```python
df=pd.DataFrame([[1,3,5],[4,5,6],[7,8,10]])

# 행 단위 최대-최소 

print(df.apply(lambda x:x.max()-x.min(),axis=0))

# 열 단위 최대-최소

print(df.apply(lambda x:x.max()-x.min(),axis=1))
```

```python
def ff(x):
    return pd.Series([x.max(),x.min(),x.max()-x.min()],index=['최대','최소','차이'])

print(df.apply(ff,axis=0))
print(df.apply(ff,axis=1))
```
